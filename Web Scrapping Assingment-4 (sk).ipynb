# 7. Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes


```python
driver=webdriver.Chrome(r'C:\Users\AnkitaKushwaha\Downloads\chromedriver.exe')
```

```python
driver.get(' https://www.imdb.com/list/ls095964455/')
```

```python
name=[]
year=[]
genre=[]
run=[]
rating=[]
voted=[]
try:
    for i in driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a'):
        name.append(i.text)
    for i in driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/span[2]'):
        year.append(i.text)
    for i in driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[5]'):
        genre.append(i.text)
    for i in driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[3]'):
        run.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]/div[1]/span[2]'):
        rating.append(i.text)
    for i in driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"][3]/span[2]'):
        voted.append(i.text)
except NoSuchElementException:
    name.append('-')
    year.append('-')
    genre.append('-')
    run.append('-')
    rating.append('-')
    voted.append('-')
```

```python
db7=pd.DataFrame({'Name':name,'Year':year,'Genre':genre,'Run Time':run,'Rating':rating,'Vote':voted})
db7
```
